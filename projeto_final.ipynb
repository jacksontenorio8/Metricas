```python
import pandas as pd
```

# Calculando métricas de avaliação de aprendizado de máquina

Neste projeto, calculamos as principais métricas para avaliação de modelos de classificação de dados:
- Acurácia
- Sensibilidade (Recall)
- Especificidade
- Precisão
- F-score

Escolhemos uma matriz de confusão arbitrária para ilustrar como cada métrica é calculada.

```python
VP = 50  # Verdadeiros Positivos
FN = 10  # Falsos Negativos
FP = 5   # Falsos Positivos
VN = 35  # Verdadeiros Negativos

matriz_confusao = pd.DataFrame(
    [[VP, FN], [FP, VN]],
    columns=["Previsto Positivo", "Previsto Negativo"],
    index=["Real Positivo", "Real Negativo"]
)
```

```python
print("Matriz de Confusão:")
display(matriz_confusao)
```

# Cálculo das métricas

```python
# Acurácia
acuracia = (VP + VN) / (VP + VN + FP + FN)
# Sensibilidade (Recall)
sensibilidade = VP / (VP + FN)
# Especificidade
especificidade = VN / (VN + FP)
# Precisão
precisao = VP / (VP + FP)
# F-score
f_score = 2 * (precisao * sensibilidade) / (precisao + sensibilidade)
```

# Exibindo os resultados


```python
print("\nMétricas Calculadas:")
print(f"Acurácia: {acuracia:.2f}")
print(f"Sensibilidade (Recall): {sensibilidade:.2f}")
print(f"Especificidade: {especificidade:.2f}")
print(f"Precisão: {precisao:.2f}")
print(f"F-score: {f_score:.2f}")
```

# Métricas:

- Acurácia: Proporção de previsões corretas em relação ao total de casos.
- Sensibilidade (Recall): Proporção de positivos corretamente identificados.
- Especificidade: Proporção de negativos corretamente identificados.
- Precisão: Proporção de positivos preditos que são realmente positivos.
- F-score: Métrica que combina precisão e sensibilidade.
